## Standard Stand-Up

Before meaningful backend work, Mike runs a short stand-up:

1. Mike defines success.
2. Sage identifies scope and reusable backend files.
3. Iris reviews user/API behavior and correction paths.
4. Blue proposes SOLID architecture and provider strategy.
5. Atlas stress-tests migrations, auth, transactions, queues, and FE/BE contracts.
6. Gauge defines diagnostics, validation responses, and log visibility.
7. Nora defines manual tests and regressions.
8. Rex confirms implementation files.
9. Vera defines verification gates.
10. Quill identifies documentation impact.

## Parallel Workflow

```text
Mike -> (Sage + Iris + Blue + Atlas + Gauge + Nora) -> Mike reconcile
     -> (Rex implementation lanes + Quill draft) -> Vera
     -> Rex fixes -> Quill final -> Mike final gate -> Oscar
```

## Mandatory Failure-State Review

Every backend flow must answer:

- What happens on validation failure?
- What happens on authorization failure?
- What happens on duplicate submit, rollback, retry, and failed sync from offline clients?
- Does the API response tell the frontend exactly what the user can correct?
- Are docs updated after Vera passes?
